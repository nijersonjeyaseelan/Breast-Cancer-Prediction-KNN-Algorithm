```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn import metrics
get_ipython().run_line_magic('matplotlib','inline')
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
print(cancer['feature_names'])
```

```python
df_feat = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df_feat.info()
```

```python
cancer['target']
```

```python
df_target = pd.DataFrame(cancer['target'],columns=['Cancer'])
print(df_target)
```

```python
df_feat.head()
```

```python
# standardizing the variables
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler(copy=True)
scaler.fit(df_feat)
```

```python
scaled_features = scaler.transform(df_feat)
```

```python
df_feat_scaled = pd.DataFrame(scaled_features,columns = df_feat.columns)
```

```python
df_feat_scaled.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(scaled_features,np.ravel(df_target),test_size=0.30,random_state = 105)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

```python
pred = knn.predict(X_test)
```

```python
print(metrics.classification_report(y_test,pred))
```

```python
# to find the best value of K 
error_rate=[]

for i in range(1,40):
    knn= KNeighborsClassifier(n_neighbors=1)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',markerfacecolor='red', markersize=10)
plt.title('Error Rate vs K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```
